# Données des loyers 

```python
import pandas as pd

# Import des datasets depuis data gouv pour les appartements 1 et 2 pièces
loyers_2024 = pd.read_csv('https://www.data.gouv.fr/fr/datasets/r/89956da9-5b9b-41d7-8703-18dbec4d54a2', encoding='latin-1', sep=';')
loyers_2023 = pd.read_csv('https://www.data.gouv.fr/fr/datasets/r/edadefbc-9707-45ef-a841-283608709e58', encoding='latin-1', sep=';')
loyers_2022 = pd.read_csv('https://www.data.gouv.fr/fr/datasets/r/7141612b-8029-44a4-a048-921a85a47b1f', encoding='latin-1', sep=';')

# Ajout d'une colonne pour l'année du fichier
loyers_2024['annee'] = 2024
loyers_2023['annee'] = 2023
loyers_2022['annee'] = 2022

# Concaténer les données des loyers
loyers_full = pd.concat([loyers_2024, loyers_2023, loyers_2022], ignore_index=True)
```

# Données des valeurs foncières

```python
# Import des données depuis data gouv
dvf_2024 = pd.read_csv('https://www.data.gouv.fr/fr/datasets/r/5ffa8553-0e8f-4622-add9-5c0b593ca1f8', compression='zip', encoding='latin-1', sep='|')
dvf_2023 = pd.read_csv('https://www.data.gouv.fr/fr/datasets/r/bc213c7c-c4d4-4385-bf1f-719573d39e90', compression='zip', encoding='latin-1', sep='|')
dvf_2022 = pd.read_csv('https://www.data.gouv.fr/fr/datasets/r/b4f43708-c5a8-4f30-80dc-7adfa1265d74', compression='zip', encoding='latin-1', sep='|')

# Ajout d'une colonne pour l'année du fichier
dvf_2024['annee'] = 2024
dvf_2023['annee'] = 2023
dvf_2022['annee'] = 2022
```

```python
# Concaténer les données des loyers
dvf_full = pd.concat([dvf_2024, dvf_2023, dvf_2022], ignore_index=True)

# Nettoyage des données
cols_to_keep = ['Date mutation', 'Nature mutation', 'Type local', 'Nombre pieces principales', 'Valeur fonciere', 'Code postal','Commune', 'Code departement', 'Code commune', '1er lot', 'Surface Carrez du 1er lot', 'annee']

dvf_filtered = dvf_full[(dvf_full['Nature mutation'].str.contains('vente', case=False)) &
                        (dvf_full['Type local'].str.contains('appartement', case=False)) &
                        (dvf_full['Nombre de lots'] == 1) &
                        (dvf_full['Nombre pieces principales'] <= 2.0)
                ]

dvf_filtered = dvf_filtered[cols_to_keep].dropna()
```
